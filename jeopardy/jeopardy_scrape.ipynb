```python
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import requests
```

```python
url = 'https://www.j-archive.com/listseasons.php'
r = requests.get(url)
soup = BeautifulSoup(r.content)
soup.prettify()
```

```python
# Get links for seasons
season_links = []
for link in soup.find_all('a'):
#     print(link.get('href'))
    if 'http' not in link.get('href'):
        season_links.append(link.get('href'))
season_links
```

```python
# Download HTML for an individual season to test code on
season_test_url = 'https://www.j-archive.com/' + season_links[5] # Just picked a random number
print(season_test_url)
r=requests.get(season_test_url)
soup = BeautifulSoup(r.content)
soup.prettify()
```

```python
# Test how to get all the links for the individual shows
episode_link_test = []
for link in soup.find_all('a'):
    if 'game_id' in link.get('href'):
        episode_link_test.append(link.get('href'))
episode_link_test
```

```python
# Loop through each of the seasons and find links to all individual episodes
episode_links = []
for i in season_links:
    season_url = 'https://www.j-archive.com/' + i
    r=requests.get(season_url)
    soup = BeautifulSoup(r.content)
    for link in soup.find_all('a'):
        if 'game_id' in link.get('href'):
            episode_links.append(link.get('href'))
print(len(episode_links))    
```
